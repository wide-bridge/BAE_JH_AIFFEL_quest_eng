```python
import random

class Account:
    account_num = 0

    def __init__(self, owner, initial_amount):
        self.bank_owner = owner
        self.bank_amount = initial_amount
        self.bank_name = "SC Bank"
        self.bank_account_num = f'{random.randint(0, 1000):03d}-{random.randint(0, 100):02d}-{random.randint(0, 1000):03d}'
        self.deposit_count = 0
        self.deposit_history_log = []
        self.withdraw_history_log = []

        Account.account_num += 1

    def get_account_num(self):
        print(f'account_num: {Account.account_num}')
        return Account.account_num

    def deposit(self, amount):
        if amount < 1:
            print('ERROR: amount sould be over 1won.')
        else:
            self.bank_amount += amount
            self.deposit_count += 1
            if self.deposit_count >= 5:
                self.bank_amount += (self.bank_amount * 0.01)
                self.deposit_count = 0
                print(f"INFO: {self.bank_owner}'s Interest has been credited.")
            self.deposit_history_log.append(f'INFO: deposit amount: {amount}  balance: {self.bank_amount}')


    def withdraw(self, amount):
        if self.bank_amount < amount:
            print(f"ERROR: {self.bank_owner}'s exceed the account balance.")
        else:
            self.bank_amount -= amount
            self.withdraw_history_log.append(f'INFO: withdraw amount: {amount}  balance: {self.bank_amount}')

    def display_info(self):
        msg = f'은행 이름: {self.bank_name}, 예금주: {self.bank_owner}, 계좌번호: {self.bank_account_num}, 잔고: {self.bank_amount:,.2f}원'
        print(msg)

    def deposit_history(self):
        print("INFO: deposit_history")
        for s in self.deposit_history_log:
            print(s)

    def withdraw_history(self):
        print("INFO: withdraw_history")
        for s in self.withdraw_history_log:
            print(s)

# -----------------------------------------------
# 초기 정보
bank_owner_lsit = ['ABC', 'DEF', 'GHI']
bank_amount_list = [10000, 20000, 30000]
account_list = []

# -----------------------------------------------
msg = '계좌 생성'
print('-' * 30, msg, '-' * 30)
for i in range(3):
    name = bank_owner_lsit[i]
    amount = bank_amount_list[i]
    account_list.append( Account(name, amount) )

# -----------------------------------------------
msg = '계좌 생성 확인 테스트'
print('-' * 30, msg, '-' * 30)
for acc in account_list:
    acc.display_info()

# -----------------------------------------------
msg = '입출금 테스트 1: 잔고 넘게 출금하기 + 로깅 기능'
print('-' * 30, msg, '-' * 30)
account_list[0].withdraw(10000)
account_list[0].withdraw(1000)

account_list[0].deposit_history()
account_list[0].withdraw_history()

# -----------------------------------------------
msg = '입출금 테스트 2 : 입금 이자 생성 확인 + 로깅 기능'
print('-' * 30, msg, '-' * 30)
account_list[1].deposit(10000)
account_list[1].deposit(10000)
account_list[1].deposit(10000)
account_list[1].deposit(10000)
account_list[1].deposit(10000)

account_list[1].deposit_history()
account_list[1].withdraw_history()
```
