**Q3. battle 함수 만들기**

    - 만약 Player 인스턴스가 살아남았다면
        - Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가

        - ‘전투 승리!’를 출력
    - Player 인스턴스가 살아남지 못했을 경우
        - ‘전투 패배..’를 출력하기

### 아래는 인간 지능으로 짠 코드

```python
import random

class Character:
    def __init__(self, name, level, health, attack, depense):
        self.name = name
        self.level = level
        self.health = health
        self.attack_power = attack
        self.depense_power = depense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        if self.depense_power < damage:
            self.health -= damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power + 1)
        target.take_damage(damage)
        msg = f"{self.name} -> {target.name} , {damage} 만큼 공격했다."
        print(msg)
        msg = f"{target.name}의 체력: {target.health}"
        print(msg)

class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack=25, depense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp

    def level_up(self):
        if self.experience >= 50:
            self.level +=  1
            self.attack_power += 10
            self.depense_power += 5

class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30 + 1) * level
        attack = random.randint(5, 20 + 1) * level
        depense = random.randint(1, 5 + 1) * level
        super().__init__(name, level, health, attack, depense)

def battle(player: Player, monster: Monster):
    is_player_wins = False

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        monster.attack_target(player)

    if player.is_alive():
        player.gain_experience(monster.level * 20)
        print("전투 승리!")
        is_player_wins = True
    else:
        print("전투 패배!")
        is_player_wins = False

    return is_player_wins

monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

name = input("input player's name: ")
player = Player(name)

monster_list = []
for name, health in monster_dict.items():
    monster_list.append( Monster(name, health))

for i in range(len(monster_list)):
    print(f"\n{monster_list[i].name}과의 전투를 시작합니다. 현재 체력: {player.health}")

    result = battle(player, monster_list[i])

    if result == False:
        print("게임 오버")
        break
    else:
        print(f"{monster_list[i].name} 를 이겼다.")
        player.gain_experience(monster_list[i].level * 20)
        player.level_up()


```

### 아래는 인공 지능으로 짠 코드

```python
import random


class Character:
    def __init__(self, name: str, level: int, hp: int, attack: int, defense: int):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self) -> bool:
        return self.hp >= 0  # 요구사항 그대로(0 이상이면 True)

    def take_damage(self, damage: int) -> None:
        real = damage - self.defense
        if real > 0:
            self.hp -= real

    def attack_target(self, target: "Character") -> int:
        dmg = random.randint(1, self.attack)
        target.take_damage(dmg)
        return dmg


class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0

    def gain_experience(self, amount: int) -> None:
        self.experience += amount

    def level_up(self) -> None:
        # 최소 구현: 경험치가 50 이상이면 1회만 레벨업하고, 경험치는 50 차감
        if self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5


class Monster(Character):
    def __init__(self, level: int):
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        dfn = random.randint(1, 5) * level
        super().__init__(name=f"Monster(Lv{level})", level=level, hp=hp, attack=atk, defense=dfn)


def battle(player: Player, monster: Monster) -> None:
    # 둘 중 하나의 hp가 0 이하가 될 때까지
    while player.hp > 0 and monster.hp > 0:
        player.attack_target(monster)
        if monster.hp <= 0:
            break
        monster.attack_target(player)

    # 결과 처리
    if player.hp > 0:
        player.gain_experience(monster.level * 20)
        player.level_up()
        print("전투 승리!")
    else:
        print("전투 패배..")


if __name__ == "__main__":
    name = input("플레이어 이름을 입력하세요: ").strip()
    player = Player(name if name else "Player")

    # 자동사냥: 몬스터 레벨 1부터 순서대로
    for lv in range(1, 101):
        if player.hp <= 0:
            break

        monster = Monster(lv)
        battle(player, monster)

    # 최소 요약 출력
    print(f"최종 상태: {player.name} (Lv{player.level}) HP={player.hp}, ATK={player.attack}, DEF={player.defense}, EXP={player.experience}")
```
