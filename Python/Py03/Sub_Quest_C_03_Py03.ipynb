```python
import math

def get_operand():
    while True:
        try:
            num = int(input('숫자를 입력하세요: '))
            return num
        except ValueError:
            print('잘못된 입력입니다. 정수를 입력하세요')
            continue

def get_operator():
    while True:
        oprt = input('연산자를 입력하세요: ')
        if oprt == '+' or oprt == '-' or oprt == "*" or oprt == "**":
            return oprt
        print("지원하지 않는 연산지 입니다.")

def get_result(a, b, oprt):

    if b == 0 and oprt == '/':
        raise Exception(ZeroDivisionError)

    result = 0
    if oprt == '+':
        result = a + b
    elif oprt == '-':
        result = a - b
    elif oprt == "*":
        result = a * b
    elif oprt == "**":
        result = math.pow(a, b)
    return result

cont = 'Y'
while cont == 'Y':
    op1 = get_operand()
    operator = get_operator()
    op2 = get_operand()

    try:
        result = get_result(op1, op2, operator)
    except ZeroDivisionError:
        print("ERROR: 0으로 나눌 수 없습니다.")
    else:
        print(f"{op1} {operator} {op2} = {result}")
    finally:
        cont = input('계속하시겠습니까?(Y/n)')

print("종료하였습니다.")
```
