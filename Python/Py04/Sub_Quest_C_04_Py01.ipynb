```python
# --------------------------------------------------------------------------------------------
# 2. 사각형 넓이 구하기
# --------------------------------------------------------------------------------------------

input_msg = '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'

class Square:
    def __init__(self):  # 생성자__init__의 매개변수의 첫번째는 self여야 합니다.
        self.square = int(input(input_msg))

        if self.square == 1:
            self.rect()
        elif self.square == 2:
            self.par()
        elif self.square == 3:
            self.trape()
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        msg = '가로, 세로를 입력하세요. 예시 : 가로, 세로\n >>>'
        width, vertical = map(int, input(msg).split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        print(result)

    def par(self):      # 평행사변형 메서드
        msg = "밑변, 높이를 입력하세요\n >>>"
        b, h = map(int, input(msg).split(','))
        result = "평행 사변형 넓이는 : " + str( b * h )
        print(result)

    def trape(self):    # 사다리꼴 메서드
        msg = "밑변, 윗변, 높이를 입력하세요\n >>>"
        b, t, h = map(int, input(msg).split(','))
        result = "사다리꼴 넓이는 : " + str((b + t) * h * 0.5)
        print(result)

square = Square()
```

```python
# --------------------------------------------------------------------------------------------
# 3. 키오스크
# --------------------------------------------------------------------------------------------

# 전체 메뉴와 가격
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price

    def menu_print(self):
        print('-' * 30)
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')
        print('-' * 30)

    def menu_select(self):
        print()
        temp = ''
        self.order_menu = []
        self.order_price = []

        while True:
            input_drink_no = 0
            #------ 메뉴 선택 하기
            while input_drink_no < 1 or len(menu) < input_drink_no:
                input_drink_no = int(input("음료 번호를 입력하세요 : "))

                if 1 <= input_drink_no & input_drink_no <= len(menu):             # 유효한 음료 번호라면
                    self.price_sum = self.price[input_drink_no-1]    # 선택 음료의 가격
                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

            #------ 온도 선택 하기
            t = 0                                       # 기본값을 넣어주고
            while t != 1 and t != 2:                    # 1이나 2를 입력할 때까지 물어보기
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                if t == 1:
                    temp = 'HOT'
                elif t == 2:
                    temp = 'ICE'
                else:
                    print("1과 2 중 하나를 입력하세요.\n")

            ordered_menu = f"{temp} {self.menu[input_drink_no-1]}"
            self.order_menu.append(ordered_menu)
            self.order_price.append(self.price[input_drink_no-1])

            msg = "추가 주문 하시겠어요? (0: 주문 없음):  "
            input_extra_order = int(input(msg))
            if input_extra_order == 0:
                break

        self.table()

    def pay(self):
        if len(self.order_menu) > 0:
            output_msg = ''
            while True:
                input_msg = "지불 수단을 선택해 주세요(cash or card or 1 or 2): "
                input_pay_method = input(input_msg)

                if input_pay_method == 'cash' or input_pay_method == '1':
                    output_msg = '직원을 호출하겠습니다.'
                    print(output_msg)
                elif input_pay_method == 'card' or input_pay_method == '2':
                    output_msg = 'IC칩 방향에 맞게 카드를 꽂아주세요.'
                    print(output_msg)
                else:
                    output_msg = '다시 결제를 시도해 주세요.'
                    print(output_msg)
                    continue
                break

    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')


a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
