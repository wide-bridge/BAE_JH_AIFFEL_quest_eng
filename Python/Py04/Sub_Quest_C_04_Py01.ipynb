```python
# --------------------------------------------------------------------------------------------
# 2. 사각형 넓이 구하기
# --------------------------------------------------------------------------------------------

input_msg = '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'

class Square:
    def __init__(self):  # 생성자__init__의 매개변수의 첫번째는 self여야 합니다.
        self.square = int(input(input_msg))

        if self.square == 1:
            self.rect()
        elif self.square == 2:
            self.par()
        elif self.square == 3:
            self.trape()
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        msg = '가로, 세로를 입력하세요. 예시 : 가로, 세로\n >>>'
        width, vertical = map(int, input(msg).split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        print(result)

    def par(self):      # 평행사변형 메서드
        msg = "밑변, 높이를 입력하세요\n >>>"
        b, h = map(int, input(msg).split(','))
        result = "평행 사변형 넓이는 : " + str( b * h )
        print(result)

    def trape(self):    # 사다리꼴 메서드
        msg = "밑변, 윗변, 높이를 입력하세요\n >>>"
        b, t, h = map(int, input(msg).split(','))
        result = "사다리꼴 넓이는 : " + str((b + t) * h * 0.5)
        print(result)

square = Square()
```
